# Leads Analysis


```python
#load data into a dataframe
import pandas as pd
df = pd.read_csv('leads_not_found.csv')
df.shape
```

```python
df.head()
```

```python
def preson_search(firstName, lastName, middleName):
    query_data = {
        "FirstName": firstName,
        "MiddleName": middleName,
        "LastName": lastName,
        "Addresses": [
            {
                "City": "",
                "State": "",
                "Zip": "",
                "AddressLine1": "",
                "AddressLine2": None,
            }
        ],
        "Dob": "",
        "Age": None,
        "AgeRange": "",
        "Phone": "",
        "Email": "",
        "Includes": ["Addresses", "PhoneNumbers"],
        "FilterOptions": ["IncludeLowQualityAddresses"],
        "Page": 1,
        "ResultsPerPage": 10,
    }
    response = requests.post(URL, json=query_data, headers=headers)
    response_data = response.json()
    list_of_persons = response_data.get("persons", [])
    
    phone_numbers = [
            item["phoneNumber"]
            for person in list_of_persons
            for item in person.get("phoneNumbers", [])
        ]
    
    return phone_numbers
    

    
```

```python

df['middle_name'] = df['middle_name'].fillna('')
df['first_name'] = df['first_name'].fillna('')
df['last_name'] = df['last_name'].fillna('')
df
```

```python
df
```

```python
df['phone_numbers_endato'] = df.apply(lambda x: preson_search(x['first_name'], x['last_name'], x['middle_name']), axis=1)
```
