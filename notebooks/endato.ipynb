## Leads Not Found Lookup

```python
today = datetime.datetime.now()

leads_not_found = get_last_lead(
    start_date=today - timedelta(days=1),
    end_date=today + timedelta(days=1),
    status="not_found",
    limit=100,
)

df = pd.DataFrame([lead.model_dump() for lead in leads_not_found])


df["middle_name"] = df["middle_name"].fillna("")
df["first_name"] = df["first_name"].fillna("")
df["last_name"] = df["last_name"].fillna("")
df["state"] = df["state"].fillna("MO")

df
```

```python
def search_person(firstName, lastName, middleName, age, city="", state="", zip="", addressLine1="", addressLine2=""):
    query_data = {
        "FirstName": firstName,
        "MiddleName": middleName,
        "LastName": lastName,
        "Addresses": [
            {
                "City": city,
                "State": state,
                "Zip": zip,
                "AddressLine1": addressLine1,
                "AddressLine2": addressLine2,
            }
        ],
        "Dob": "",
        "Age": age,
        "AgeRange": "",
        "Includes": ["PhoneNumbers", "EmailAddresses"],
        "FilterOptions": [
            "EnableExactFirstNameMatch",
            "EnableExactLastNameMatch",
            "DisableFirstNameSynonymMatching",
            "DisableSmartSearch",
        ],
        "Page": 1,
        "ResultsPerPage": 1,
    }
    print(query_data)
    response = requests.post(URL, json=query_data, headers=headers)
    response_data = response.json()
    list_of_persons = response_data.get("persons", [])

    if len(list_of_persons) > 1:
        print("More than one person found")
        print("selecting the first person")
        return list_of_persons[0]
    elif len(list_of_persons) == 0:
        print("No person found")
        return None

    return list_of_persons[0]

    phone_numbers = [
            item["phoneNumber"]
            for person in list_of_persons
            for item in person.get("phoneNumbers", [])
        ]

    return phone_numbers
```

```python
results = {}

for index, row in df.iterrows():
    print(f"Looking for {row['case_id']}")
    print(row['first_name'], row['last_name'], row['middle_name'], row['age'], row['city'], row['state'], row['zip_code'], row['address'])
    results[row['case_id']] = search_person(firstName=row['first_name'], lastName=row['last_name'], middleName=row['middle_name'], age=row['age'], city=row['city'], state=row['state'], zip=row['zip_code'], addressLine1=row['address'])
    break
```
